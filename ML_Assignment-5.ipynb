```python
# Dataset:Use the Iris dataset available in the sklearn library.
# Load the Iris dataset from sklearn. Drop the species column since this is a clustering problem.
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
# KMeans Clustering

# How K-Means Clustering Works:
      # K-Means is a popular clustering algorithm that aims to partition data into ùëò clusters. 

# It works by:
     # .Initializing ùëò centroids randomly.
     # .Assigning each data point to the nearest centroid.
     # .Updating centroids based on the mean of assigned points.
     # .Repeating steps 2 and 3 until centroids stabilize.

# Why K-Means is Suitable for the Iris Dataset:
       # .K-Means efficiently groups data based on numerical features like petal and sepal measurements.
       # .Since the dataset is small and relatively balanced, K-Means performs well in detecting natural groupings.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)  # Applying kMeans
df['cluster'] = kmeans.fit_predict(df)

plt.figure(figsize=(8, 6))  # Visualizing cluster
sns.scatterplot(x=df.iloc[:, 2], y=df.iloc[:, 3], hue=df['cluster'], palette="viridis")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-Means Clustering of Iris Dataset")
plt.legend(title="Cluster")
plt.show()
```

```python
# B) Hierarchical Clustering

# How Hierarchical Clustering Works:
      # Hierarchical clustering builds a hierarchy of clusters in two ways:

   # Agglomerative (bottom-up): Each point starts as its own cluster, and similar clusters merge iteratively.
   # Divisive (top-down): All points start in one cluster and split recursively into smaller clusters.

# Why Hierarchical Clustering is Suitable for the Iris Dataset:
     # .The method captures natural hierarchies in the data.
     # .No need to specify the number of clusters beforehand.
     # .Can visualize relationships between clusters using dendrograms.
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

linked = linkage(df.iloc[:, :4], method='ward')   # Compute hierarchical clustering

# Plot Dendogram
plt.figure(figsize=(20, 10))   # Visualizing cluster
dendrogram(linked, labels=np.arange(len(df)), leaf_rotation=90, leaf_font_size=8)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Point Index")
plt.ylabel("Cluster Distance")
plt.show()
```
